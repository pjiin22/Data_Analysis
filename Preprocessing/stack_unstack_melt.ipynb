## **stack, unstack**
 - stack : 컬럼 레벨에서 인덱스 레벨로 데이터프레임을 변경합니다.
 - unstack : 인덱스 레벨에서 컬럼 레벨로 데이터프레임을 변경합니다.
  - stack의 반대입니다.

```python
import pandas as pd

df = pd.read_csv("titanic_train.csv")
```

```python
df.head()
```

```python
pivot = pd.pivot_table(df, index=['Sex','Pclass'], values=['Survived','Fare'], aggfunc=['mean','median','sum'])
pivot
```

```python
pivot.stack(0) 
#컬럼의 첫번째 레벨을 인덱스로 내립니다.
#컬럼의 첫 번째 레벨(위쪽의 멀티인덱스가 있을 경우 : 위에부터 첫번째)을 인덱스로 "내린다"는 뜻
```

```python
pivot.stack(1) #컬럼의 두번째 레벨을 인덱스로 내립니다.
```

```python
pivot.unstack(0) #인덱스의 첫번째 레벨을 컬럼으로 쌓아 올립니다
```

```python
pivot.unstack(1) #인덱스의 두번째 레벨을 컬럼으로 쌓아 올립니다
```

## **melt**
- `pd.melt(데이터명, id_vars=기준 컬럼)`
![이미지](https://i0.wp.com/cmdlinetips.com/wp-content/uploads/2019/06/Pandas_melt_reshape.png?w=421&ssl=1)

```python
data = pd.DataFrame({'name':['a','b','c']
                    , 'order_count':[3,4,10]
                    , 'amount':[10000,25000,300000]})
```

```python
data
```

```python
pd.melt(data, id_vars=['name'])
#name은 열로 유지하고 나머지는 (인덱스로만 쳤을 때)행 방향에서 열 방향으로 바꾼다
```

```python
pd.melt(data, id_vars=['name'], var_name='type', value_name='val')
#var_name=' ': 열이름이 변환되어 내려갈 때 그 컬럼의 이름 지정
#value_name=' ': 기존 값이 변환되어 내려갈 때 그 컬럼의 이름 지정
```
