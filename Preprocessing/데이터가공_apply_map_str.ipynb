## **apply 함수**
- 사용자 정의 함수를 데이터에 적용하고 싶을 때 사용합니다.
- `.apply(함수, axis=0/1)`
- axis=1 : 각 행을 함수에 전달 & aixs=0 : 각 열을 함수에 전달

```python
df1 = df.copy()
```

```python
def pclass_sibsp(x):
    if x['Pclass'] == 1 and x['SibSp'] == 1:
        return 1
    else:
        return 0
```

```python
df1['pclass_sibsp_filter'] = df1.apply(pclass_sibsp, axis=1) #def 함수를 쓰면 꼭 apply를 쓴다는 것!

#df1의 행(axis = 1)에 pclass_sibsp를 적용시킨 결과를 pclass_sibsp_filter 열에 넣어라
#행의 pclass가 1이고 sibsp가 1이면 1을 return, 아니면 0을 return
```

```python
df1.head()
```

```python
import numpy as np

def adult(x):
    if x >= 19:
        return 1
    elif x < 19:
        return 0
    else:
        return np.nan
```

```python
df1['adult_yn'] = df1['Age'].apply(adult) #특정 컬럼에 adult 함수를 적용시키는 것
df1.head(10)
```

- 간단한 함수는 lambda로 구현할 수 있습니다.

```python
df1['pclass_sibsp_lambda'] = df1.apply(lambda x: 1 if x['Pclass'] == 1 and x['SibSp'] == 1 else 0, axis=1) #매개변수는 x이고 표현식은 그 뒤
df1.head()
```

## **map 함수**
- 값을 특정 값으로 치환하고 싶을 때 사용합니다.
- `데이터명[컬럼명].map(매핑 딕셔너리)`

```python
gender_map = {'male':'남자', 'female':'여자'}
df1['Sex_kr'] = df1['Sex'].map(gender_map) #df1의 'sex'에 gender_map을 적용시켜서 값을 변환시킨 후 그 값을 sex_kr에 넣기
df1.head()
```

## **문자열 다루기**
|메소드|설명|
|:--|:--|
|`.str.contains(문자열)`|문자열을 포함하고 있는지 유무|
|`.str.replace(기존문자열, 대치문자열)`|문자열 대치|
|`.str.split(문자열, expand=True/False, n=개수)`|특정 문자열을 기준으로 쪼개기|
|`.str.lower()`|소문자로 바꾸기|
|`.str.upper()`|대문자로 바꾸기|

```python
df2 = df.copy()
df2.head()
```

- `.str.contains(문자열)`

```python
df2['Name'].str.contains('Mrs') #문자열을 포함하고있는가
```

```python
df2[df2['Name'].str.contains('Mrs')] #Mrs를 포함하는 행들만 필터링하여 '새로운 데이터프레임'을 반환
```

- `.str.replace(기존문자열, 대치문자열)`

```python
df2['Name'] = df2['Name'].str.replace(',', '') #문자열 대치
df2.head()
```

- `.str.split(문자열, expand=True/False, n=개수)`

```python
df2['Name'].str.split(' ') #특정 문자를 기준으로 split하기
```

```python
df2['Name'].str.split(' ', expand=True, n=1)
# expand=True: 분할된 결과를 데이터프레임의 여러 열로 반환
# expand=False: 분할된 결과가 리스트로 반환되고, 그 리스트가 데이터프레임의 한 열에 저장됩니다.

# n=1: 첫 번째 공백까지만 분할하고, 그 이후는 분할하지 않고 남김 -> 이 경우 expand = True여서 다른 열에 저장
```

- `.str.lower()`
- `.str.upper()`

```python
df2['Name'].str.lower() #소문자로 바꾸기
```

```python
df2['Name'].str.upper() #대문자로 바꾸기
```
