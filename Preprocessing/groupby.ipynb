**groupby**
- 같은 값을 한 그룹으로 묶어서 여러 가지 연산 및 통계를 구할 수 있습니다.
- `데이터.groupby(컬럼명).연산및통계함수`

---



```python
import pandas as pd

df = pd.read_csv("/content/titanic_train.csv")
```

```python
df.head()
```

```python
df['PassengerId'] = df['PassengerId'].astype(str) #타입변경
```

## **단일 그룹**
|함수|설명|
|:--|:--|
|count()|행의 갯수|
|nunique()|행의 유니크한 갯수|
|sum()|합|
|mean()|평균|
|min()|최솟값|
|max()|최댓값|
|std()|표준편차|
|var()|분산|

```python
df.groupby('Pclass').count() #group으로 묶은 후에 개수 세기
```

```python
df.groupby('Pclass').nunique() #유니크한 갯수
```

```python
df.groupby('Pclass').sum(numeric_only=True) #묶은 후에 sum을 하는데 숫자인 값만 계산
```

```python
df.groupby('Pclass').mean(numeric_only=True)
```

```python
df.groupby('Pclass').min(numeric_only=True)
```

```python
df.groupby('Pclass').max(numeric_only=True)
```

```python
df.groupby('Pclass').std(numeric_only=True)
```

```python
df.groupby('Pclass').var(numeric_only=True)
```

```python
df.groupby('Pclass')[['Survived']].mean() #묶은 후에 특정 컬럼만 mean해주기
```

```python
df.groupby('Pclass')[['Survived','Age']].mean()
```

## **다중 그룹**

```python
df.groupby(['Sex','Pclass']).mean(numeric_only=True) #두개의 그룹으로 묶기 -> []로 묶어주기
```

```python
import numpy as np

df.groupby(['Sex','Pclass'])[['Survived','Age','SibSp','Parch','Fare']].aggregate([np.mean, np.min, np.max])

# 여러 통계 함수를 동시에 적용하고, 특정 열에 대해서만 다양한 통계값을 계산
##Survived, Age, SibSp, Parch, Fare 열에 대해서만 통계 함수를 적용
##aggregate([np.mean, np.min, np.max])는 각 그룹에 대해 평균(mean), 최소값(min), **최대값(max)**을 계산하도록 지정
```
